```python
#get titles
def dataset_gen(directory):
    for name in os.listdir(directory):
        full_path = os.path.join(directory, name)
        if os.path.isfile(full_path):
            yield os.path.splitext(name)[0]
        else:
            print('Unidentified name %s. It could be a symbolic link' % full_path)
```

```python
def get_title(element):
    return plaintext(element.by_class("recipe-summary__h1")[0].content)
```

```python
dataset = dataset_gen("web_data/img/")
```

```python
titles = {}
for idx in dataset:
    print(idx)
    url = URL("http://allrecipes.com/recipe/"+str(idx))
    html = url.download(cached=True)
    element = Element(html)
    title = get_title(element)
    titles[idx]=title
    print(title)
    
```
